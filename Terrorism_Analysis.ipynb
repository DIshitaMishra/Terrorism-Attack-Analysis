#Exploratory Data Analysis (EDA):

```python
# What is the overall trend of terrorism attacks over the years?
# Has the frequency of attacks changed over the years?

attacks_by_year = df['year'].value_counts().sort_index()
attacks_by_year
```

```python
# Plotting the trend of attacks over the years

plt.plot(attacks_by_year.index, attacks_by_year.values, marker = '.',color = 'red')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title('Trend of Terrorism Attacks Over the Years')
plt.show()
```

"The trend in terrorism attacks exhibits distinct phases. Starting from a low point in 1970, attacks increased steadily until the early 1990s. Subsequently, there was a declining trend, followed by a resurgence in the mid-2000s. However, the most significant shift occurred after 2011, with a sharp decrease in attacks. These trends reflect the dynamic and evolving nature of terrorism."

```python
# Which countries or regions are most affected by terrorism?

region = df.groupby('region_txt')['eventid'].count().sort_values(ascending=False)
region
```

```python
# Create a bar chart

plt.figure(figsize=(12,6))
region.head(10).plot(kind = 'bar', color = 'skyblue')
plt.title("Top 10 Regions Most Affected by Terrorism")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
nation = df.groupby('nationlity_txt')['eventid'].count().sort_values(ascending=False)
nation
```

```python
# top 10 Country with highest attacks 
plt.figure(figsize=(12,6))
nation.head(10).plot(kind = 'bar', color = 'green')
plt.title("Top 10 Nation Most Affected by Terrorism")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

Different Number of attacks in the different region and differnt Country. Most affected region is Middle east & North Africa and South Asia whereas most affected countries are Iraq, Pakistan and so on given in above graphs

```python
# What are the most common types of attacks (e.g., bombings, shootings)?

attack = df['attacktype1_txt'].value_counts()
plt.figure(figsize=(12,6))
attack.plot(kind = 'barh', color = 'lightcoral')
plt.xlabel('Attack type')
plt.title('Most Common Types of Terrorist Attacks')
plt.ylabel('Number of Attacks')
plt.show()
```

"In the realm of terrorism, bombings, explosions, and armed assaults stand out as the most frequently employed methods of attack. These trends highlight the enduring prevalence of violent actions involving explosives and firearms. Recognizing these patterns is essential for assessing the ever-evolving landscape of terrorism and crafting strategic responses."

```python
# What are the primary targets of these attacks (e.g., civilians, military, government)?

target =  df['targtype_txt'].value_counts()

plt.figure(figsize=(12,6))
target.plot(kind = 'bar', color = 'lightgreen')
plt.title('Primary Targets of Terrorist Attacks')
plt.xlabel('Target Type')
plt.ylabel('Number of Attacks')
plt.show()
```

"Private citizens and their property are the primary targets of terrorist attacks, with a notably higher frequency of incidents. This underscores the urgency of protecting civilian lives and assets in counterterrorism efforts."

```python
# Are there any noticeable patterns or trends in the timing of attacks (e.g., seasonality)?
# Are there specific months when attacks are more common?

attcks_month = df.groupby(['year','month']).size().unstack(fill_value=0).stack()

plt.figure(figsize=(12,6))
attcks_month.plot(kind = 'line', color = 'lightseagreen')
plt.title('Trends in Monthly Terrorist Attacks')
plt.xlabel('Time (Year-Month)')
plt.ylabel('Number of Attacks')
plt.show()
```

The above figure shows the trends in monthly terrorist attacks 

```python
# What is the rate of success and failure for these attacks?

# Group data by success and calculate the total count
success_count = df['success'].sum()
failure_count = len(df) - success_count

# Calculate success and failure rates
total_attacks = len(df)
success_rate = success_count / total_attacks
failure_rate = 1 - success_rate
print(success_rate)
print(failure_rate)
```

```python

# Create a pie chart to visualize success and failure rates
labels = ['Success', 'Failure']
sizes = [success_rate, failure_rate]
colors = ['lightgreen', 'lightcoral']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Success and Failure Rates of Terrorist Attacks')
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart

plt.show()
```

"An overwhelming 91% of terrorist attacks achieve success, with only 8% resulting in failure. This underscores the high effectiveness of these acts, necessitating strong counterterrorism strategies."
